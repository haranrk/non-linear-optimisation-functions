def descent_update_AG(x, desc_dir, alpha=0.25, beta=0.5):
    """
    x: is a d-dimensional vector giving the current represented as numpy array.
    desc_dir: is d-dimensional vector giving the descent direction, the next iterate
    is along this direction.
    Returns
    Next iterate where step size is chosen to satisfy Armijo-Goldstein conditions.
    """
    grad = get_gradient(x)
    val = get_value(x)
    grad_component = np.dot(grad, desc_dir)
    if grad_component >=0 :
      print("ERROR : NOT A DESCENT DIRECTION!!")
      return -1
    cand_eta = 1
    eta_min = 0 # the smallest eta to try
    eta_max = np.inf # the largest eta to try
    while True:
        cand_x = x + cand_eta * desc_dir
        f = get_value(cand_x)
        f1 = val + beta * cand_eta * grad_component
        f2 = val + alpha * cand_eta * grad_component # Note f2 is always greater than f1.
        if f>=f1 and f<=f2:
            # value at point is between both lines, so stop
            break
        if f<f1:
            # value at point is below both lines, so increase eta,
            # only try eta larger than current eta
            eta_min = cand_eta
            cand_eta = np.min([2*cand_eta, (cand_eta+eta_max)/2])
            # increasing eta to mid point of current eta and max eta.
            # Setting to 2*current eta, if max eta =infinity.
        if f>f2:
            # value at point is above both lines, so decrease eta,
            # only try eta smaller than current eta
            eta_max = cand_eta
            cand_eta = (cand_eta+eta_min)/2
            # decreasing eta to mid point of current eta and min eta.
    return cand_x
